```python
import datetime
import time

start_date = "8/27/18"
start = datetime.datetime.strptime(start_date, "%m/%d/%y")

current = start

for d in range(45):
    if (time.strftime("%A", current.timetuple()) == "Monday"):
        print("<tr>\n<th>Week 1</th>")
    
    print("<td>\n<h4>" + time.strftime("%A, %B %e", current.timetuple()) + "</h4>\n\n</td>")
    
    if (time.strftime("%A", current.timetuple()) == "Friday"):
        print("</tr>\n")
    
    if (current.weekday() is 4):
        current = current + datetime.timedelta(days=3)
    else:
        current = current + datetime.timedelta(days=2)
```
